```python
df1 = pd.read_csv('japan_housing\FEH_00200522_240325173059.csv')
df1.head()
```

```python
df2 = pd.read_csv('japan_housing\FEH_00200522_240325173122.csv')
df2.head()
```

```python
japan_housing = pd.concat([df, df1, df2], axis=0)
japan_housing.head()
```

```python
#各データ数分読み取れているか確認。行数は275100になるはず。
japan_housing.shape
```

```python
#コードが邪魔な気がするので消す。でも結局質的変数なのには変わらないからいるのだろうか。よくわからないからとりあえず消してみる。
#結局地域コードは中央区とかだと他地方にもありそうなので残すことにした。これでいいのかは分からない。
japan_housing = japan_housing.drop([ '時間軸(年次) コード', '時間軸(年次)', '世帯の種類 コード', '住宅の所有の関係 コード', '世帯の年間収入階級 コード', '/表章項目'], axis=1)
japan_housing.head()
```

```python
japan_housing.shape
```

```python
#全国の行数210。(全国＋都道府県)*210=48*210=10080と一致するかどうか。
nationwide_regions = ['全国', '北海道', '青森県', '岩手県', '宮城県', '秋田県', '山形県', '福島県', '茨城県', '栃木県', '群馬県', '埼玉県', '千葉県', '東京都', '神奈川県', '新潟県', '富山県', '石川県', '福井県', '山梨県', '長野県', '岐阜県', '静岡県', '愛知県', '三重県', '滋賀県', '京都府', '大阪府', '兵庫県', '奈良県', '和歌山県', '鳥取県', '島根県', '岡山県', '広島県', '山口県', '徳島県', '香川県', '愛媛県', '高知県', '福岡県', '佐賀県', '長崎県', '熊本県', '大分県', '宮崎県', '鹿児島県', '沖縄県']
filtered_japan_housing = japan_housing[japan_housing['地域'].isin(nationwide_regions)]
filtered_japan_housing.shape
```

```python
japan_housing = japan_housing[~japan_housing['地域'].isin(nationwide_regions)]
japan_housing.head()
```

```python
#3つのデータを統合した総行数275100から、全国＋都道府県の行数10080を引いた数265020と一致するかどうか。
japan_housing.shape
```

```python
japan_housing = japan_housing[~japan_housing['世帯の種類'].str.contains('総数')]
japan_housing = japan_housing[~japan_housing['住宅の所有の関係'].str.contains('総数')]
japan_housing = japan_housing[~japan_housing['世帯の年間収入階級'].str.contains('総数')]
```

```python
japan_housing.shape
```

```python
japan_housing.head()
```

```python
#世帯の年間収入階級の階級値を中央値に変換
japan_housing = japan_housing.replace('100万円未満', '50')
japan_housing = japan_housing.replace('100～200万円未満', '150')
japan_housing = japan_housing.replace('200～300万円未満', '250')
japan_housing = japan_housing.replace('300～400万円未満', '350')
japan_housing = japan_housing.replace('400～500万円未満', '450')
japan_housing = japan_housing.replace('500～700万円未満', '600')
japan_housing = japan_housing.replace('700～1000万円未満', '850')
japan_housing = japan_housing.replace('1000～1500万円未満', '1250')
japan_housing = japan_housing.replace('1500万円以上', '1500')
japan_housing.head()
```

```python
japan_housing = japan_housing.replace('主世帯', '1')
japan_housing = japan_housing.replace('同居世帯・住宅以外の建物に居住する世帯', '2')
japan_housing = japan_housing.replace('持ち家', '1')
japan_housing = japan_housing.replace('借家', '2')
japan_housing = japan_housing.replace('公営の借家', '21')
japan_housing = japan_housing.replace('都市再生機構(UR)・公社の借家', '22')
japan_housing = japan_housing.replace('民営借家', '23')
japan_housing = japan_housing.replace('給与住宅', '24')
japan_housing.head()
```

```python
japan_housing = japan_housing.drop([ '地域'], axis=1)
japan_housing.head()
```

```python
japan_housing.shape
japan_housing.dtypes
```

```python
#欠損値と表記の揺れの修正
japan_housing.replace('***', 0, inplace=True)
japan_housing.replace('-', 0, inplace=True)
japan_housing['世帯の種類'] = japan_housing['世帯の種類'].astype(int)
japan_housing['住宅の所有の関係'] = japan_housing['住宅の所有の関係'].astype(int)
japan_housing['世帯の年間収入階級'] = japan_housing['世帯の年間収入階級'].astype(int)
japan_housing['普通世帯数【世帯】'] = japan_housing['普通世帯数【世帯】'].str.replace(',', '')
japan_housing['普通世帯数【世帯】'] = japan_housing['普通世帯数【世帯】'].fillna('0')
japan_housing['普通世帯数【世帯】'] = japan_housing['普通世帯数【世帯】'].astype(int)
japan_housing['１世帯当たり人員【人】'] = japan_housing['１世帯当たり人員【人】'].astype(float)
japan_housing['１世帯当たり居住室数【室】'] = japan_housing['１世帯当たり居住室数【室】'].astype(float)
japan_housing['１世帯当たり居住室の畳数【畳】'] = japan_housing['１世帯当たり居住室の畳数【畳】'].astype(float)
japan_housing.head()
```
